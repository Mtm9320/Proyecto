# Importar datos

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.cm import get_cmap
from matplotlib.cm import viridis
import matplotlib.cm as cm
```

```python
pd.set_option('display.max_columns', None)
```

```python
promesas = pd.read_csv("promesas.csv")
promesas.head()
```

```python
promesas.rename(columns={'med_cen_def_izq': 'med_cen_of_izq', 'med_cen_of_izq': 'med_cen_of_dch'})
```

# Valor de mercado por nacionalidad y ao

```python
nacionalidades_seleccionadas = ["Inglaterra", "Espana", "Italia", "Francia", "Brasil"]
datos_filtrados = promesas[promesas['nacionalidad'].isin(nacionalidades_seleccionadas)]

# Agrupar por 'nacionalidad' y 'version_x' y sumar 'valor_de_mercado'
datos_agrupados_por_version = datos_filtrados.groupby(
    ['nacionalidad', 'version_x']
)['valor_de_mercado'].sum().reset_index()

# Crear una tabla pivote
datos_pivoteados = datos_agrupados_por_version.pivot(
    index='version_x',
    columns='nacionalidad',
    values='valor_de_mercado'
).reindex(columns=nacionalidades_seleccionadas)

# Generar colores diferenciados dentro de la paleta 'viridis'
colores_variados = cm.viridis(np.linspace(0.2, 0.8, len(datos_pivoteados.columns)))

# Graficar un grfico de lneas para cada nacionalidad con colores ligeramente variados
plt.figure(figsize=(10, 5))

for i, nacionalidad in enumerate(datos_pivoteados.columns):
    plt.plot(
        datos_pivoteados.index, 
        datos_pivoteados[nacionalidad], 
        marker='o', 
        label=nacionalidad, 
        color=colores_variados[i]
    )

plt.title('Valor de Mercado Total por Nacionalidad y Temporada', fontsize=14)
plt.xlabel('Temporada', fontsize=12)
plt.ylabel('Valor de Mercado (Suma)', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Nacionalidad', bbox_to_anchor=(1, 1), loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# Salario semanal por nacionalidad y ao

```python
nacionalidades_seleccionadas = ["Inglaterra", "Espana", "Italia", "Francia", "Brasil"]
datos_filtrados = promesas[promesas['nacionalidad'].isin(nacionalidades_seleccionadas)]

# Agrupar los datos por nacionalidad y temporada, sumando el salario semanal
datos_agrupados_salario = datos_filtrados.groupby(['nacionalidad', 'version_x'])['salario_semanal'].sum().reset_index()

# Crear una tabla pivote basada en los salarios
datos_pivoteados_salario = datos_agrupados_salario.pivot(
    index='version_x', 
    columns='nacionalidad', 
    values='salario_semanal'
).reindex(columns=nacionalidades_seleccionadas)

# Generar colores diferenciados dentro de la paleta 'viridis'
colores_variados = cm.viridis(np.linspace(0.2, 0.8, len(datos_pivoteados_salario.columns)))

# Graficar un grfico de lneas para cada nacionalidad con colores ligeramente variados
plt.figure(figsize=(10, 5))

for i, nacionalidad in enumerate(datos_pivoteados_salario.columns):
    plt.plot(
        datos_pivoteados_salario.index, 
        datos_pivoteados_salario[nacionalidad], 
        marker='o', 
        label=nacionalidad, 
        color=colores_variados[i]
    )

plt.title('Salario Semanal Total por Nacionalidad y Temporada', fontsize=14)
plt.xlabel('Temporada', fontsize=12)
plt.ylabel('Salario Semanal (Suma)', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Nacionalidad', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# Promesas menores de 26 aos por liga

```python
ligas_seleccionadas_actualizadas = ["Premier League", "La Liga", "Serie A", "Bundesliga", "Ligue 1"]

# Filtrar jugadores menores de 26 aos y pertenecientes a las ligas seleccionadas
jugadores_filtrados_actualizados = promesas[
    (promesas['edad'] < 26) & 
    (promesas['nombre_liga_x'].isin(ligas_seleccionadas_actualizadas))
]

# Contar la cantidad de jugadores por liga
conteo_jugadores_actualizado = jugadores_filtrados_actualizados['nombre_liga_x'].value_counts()

# Generar colores con la paleta 'viridis'
colores_viridis = cm.viridis(np.linspace(0.2, 0.8, len(conteo_jugadores_actualizado)))

# Graficar el resultado
plt.figure(figsize=(10, 5))
conteo_jugadores_actualizado.plot(
    kind='bar', 
    color=colores_viridis
)

plt.title('Cantidad de promesas menores de 26 aos por Liga', fontsize=14)
plt.xlabel('Liga', fontsize=12)
plt.ylabel('Cantidad de Jugadores', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# Promesas menores de 26 aos por temp.

```python
ligas_seleccionadas_actualizadas = ["Premier League", "La Liga", "Serie A", "Bundesliga", "Ligue 1"]

# Filtrar jugadores menores de 26 aos y pertenecientes a las ligas seleccionadas
jugadores_filtrados_actualizados = promesas[
    (promesas['edad'] < 26) & 
    (promesas['nombre_liga_x'].isin(ligas_seleccionadas_actualizadas))
]

# Agrupar los datos por 'version_x' (temporada) y 'nombre_liga_x',
# contando la cantidad de jugadores por temporada en cada liga
conteo_jugadores_por_version = jugadores_filtrados_actualizados.groupby(
    ['nombre_liga_x', 'version_x']
).size().reset_index(name='cantidad_jugadores')

# Reorganizar los datos en una tabla pivote para facilitar la visualizacin
datos_pivoteados_version = conteo_jugadores_por_version.pivot(
    index='version_x', columns='nombre_liga_x', values='cantidad_jugadores'
).reindex(columns=ligas_seleccionadas_actualizadas)

# Generar colores con la paleta 'viridis' para las barras
colores_viridis_barras = cm.viridis(np.linspace(0.2, 0.8, len(ligas_seleccionadas_actualizadas)))

# Graficar un grfico de barras agrupadas por temporada para cada liga con colores 'viridis'
datos_pivoteados_version.plot(
    kind='bar',
    figsize=(10, 5),
    color=colores_viridis_barras
)

plt.title('Cantidad de promesas menores de 26 aos por Liga y Temporada', fontsize=14)
plt.xlabel('Temporada', fontsize=12)
plt.ylabel('Cantidad de Jugadores', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Liga', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# Promedio de edad

```python
# Calcular la edad promedio para cada categora
summary = promesas.groupby('categoria')['edad'].mean().reset_index()
summary.rename(columns={'edad': 'average_age'}, inplace=True)

# Generar colores basados en la paleta 'viridis'
unique_categories = promesas['categoria'].unique()
colormap = cm.viridis(np.linspace(0.2, 0.8, len(unique_categories)))
color_mapping = {category: colormap[i] for i, category in enumerate(unique_categories)}

# Configurar el grfico
plt.figure(figsize=(10, 5))

# Histograma de edades segmentado por categora con colores 'viridis'
sns.histplot(
    data=promesas, 
    x='edad', 
    hue='categoria', 
    element='step', 
    bins=20, 
    kde=False, 
    palette=color_mapping
)

# Agregar lneas verticales para la edad promedio
for _, row in summary.iterrows():
    plt.axvline(x=row['average_age'], color='red', linestyle='--', linewidth=1.5)
    plt.text(row['average_age'] + 0.3, 50, f"{row['average_age']:.2f}", color='red')

# Finalizar el grfico
plt.title("Distribucin y Edad Promedio de Promesas")
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.grid(visible=False)
plt.tight_layout()
plt.show()
```

```python
position_to_column = {
    'lw': 'extr_izq',
    'lf': 'seg_del_izq',
    'cf': 'media_punta',
    'rf': 'seg_del_dch',
    'rw': 'extr_dch',
    'ls': 'del_izq',
    'st': 'del',
    'rs': 'del_dch',
    'lam': 'med_cen_of_izq',
    'cam': 'med_cen_of',
    'ram': 'med_cen_of_dch',
    'lm': 'med_izq',
    'lcm': 'med_cen_izq',
    'cm': 'med_cen',
    'rcm': 'med_cen_dch',
    'rm': 'med_dch',
    'ldm': 'med_def_izq',
    'cdm': 'med_cen_def',
    'rdm': 'med_cen_def_dch',
    'lwb': 'carr_izq',
    'lb': 'lat_izq',
    'lcb': 'def_cen_izq',
    'cb': 'def_cen',
    'rcb': 'def_cen_dch',
    'rb': 'lat_dch',
    'rwb': 'carr_dch',
    'gk': 'portero'
}
position_labels = {
    'GK': 'POR','RWB': 'CAD','RB': 'LAD','CB': 'CEN','RCB': 'CDE','LCB': 'CIZ','LB': 'LIZ','LWB': 'CAI','RM': 'MED',
    'CM': 'MEC','LM': 'MEI','CDM': 'MCD','CAM': 'MCO','RF': 'SDD','CF': 'MPN','LF': 'SDI','RW': 'EXD','ST': 'DC',
    'LW': 'EXI','LS': 'DEI','RS': 'DED','LAM': 'MOI','RAM': 'MOD','LCM': 'MCI','RCM': 'MCD','LDM': 'MDI','RDM': 'MDD'}

def plot_player_positions_with_grass_background_only(player_name, selected_version, promesas):
    # Normalizar los nombres para mejorar la coincidencia
    promesas['nombre_normalizado'] = promesas['nombre_jugador'].str.lower()
    player_name_normalized = player_name.lower()

    # Filtrar jugadores que coincidan con el nombre
    name_parts = player_name_normalized.split()
    player_data = promesas[
        promesas['nombre_normalizado'].apply(lambda x: all(part in x for part in name_parts))
    ]
    
    if player_data.empty:
        print(f"No se encontr ningn jugador con coincidencia para '{player_name}'.")
        return

    # Mostrar las versiones disponibles solo para este jugador
    available_versions = player_data['version_x'].unique()
    print(f"Versiones disponibles para '{player_name}':")
    for version in available_versions:
        team = player_data[player_data['version_x'] == version]['nombre_club_y'].iloc[0]
        print(f"- {version} (Equipo: {team})")
    
    if str(selected_version) not in map(str, available_versions):
        print(f"La versin seleccionada '{selected_version}' no est disponible para '{player_name}'.")
        return

    # Filtrar el jugador por la versin seleccionada
    player_data = player_data[player_data['version_x'].astype(str) == str(selected_version)]
    
    # Usar el primer jugador encontrado
    best_match = player_data.iloc[0]['nombre_jugador']
    selected_team = player_data.iloc[0]['nombre_club_y'].title()
    print(f"Jugador encontrado: {best_match}, Versin: {selected_version}, Equipo: {selected_team}")
    
    # Extraer las posiciones y calificaciones correspondientes
    position_ratings = {}
    for pos, col in position_to_column.items():
        if col in player_data.columns:
            position_ratings[pos] = int(player_data.iloc[0][col]) if not pd.isna(player_data.iloc[0][col]) else 0
        else:
            position_ratings[pos] = 0  # Valor por defecto si la columna no existe

    # Crear el grfico de posiciones
    fig, ax = plt.subplots(figsize=(5, 5))
    fig.patch.set_facecolor('#2E8B57')  # Verde csped
    ax.set_facecolor('#2E8B57')  # Verde csped
    ax.set_xlim(0, 7)
    ax.set_ylim(0, 5)
    ax.axis('off')

    # Dibujar las lneas del campo de ftbol
    field_lines = [
        [(0, 0), (7, 0)],  # Lnea de gol inferior
        [(0, 5), (7, 5)],  # Lnea de gol superior
        [(0, 0), (0, 5)],  # Lnea lateral izquierda
        [(7, 0), (7, 5)],  # Lnea lateral derecha
        [(3.5, 0), (3.5, 5)],  # Lnea central
        [(1.5, 0), (1.5, 1)],  # rea de penal izquierda
        [(5.5, 0), (5.5, 1)],  # rea de penal derecha
        [(1.5, 4), (1.5, 5)],  # rea de penal superior izquierda
        [(5.5, 4), (5.5, 5)],  # rea de penal superior derecha
        [(1.5, 1), (5.5, 1)],  # Lnea de penal inferior
        [(1.5, 4), (5.5, 4)],  # Lnea de penal superior
    ]
    for line in field_lines:
        ax.plot(*zip(*line), color='white', linewidth=1.5)

    # Dibujar el crculo central
    center_circle = plt.Circle((3.5, 2.5), 0.75, color='white', fill=False, linewidth=1.5)
    ax.add_patch(center_circle)

    # Colores basados en las calificaciones
    position_grid = {
        'lw': (0.5, 4), 'lf': (1.5, 4), 'cf': (3.5, 3.5), 'rf': (5.5, 4), 'rw': (6.5, 4),
        'ls': (1, 3.5), 'st': (3.5, 4), 'rs': (6, 3.5),
         'cam': (3.5, 3), 
        'lm': (0.5, 2.5), 'lcm': (2, 2.5), 'cm': (3.5, 2.5), 'rcm': (5, 2.5), 'rm': (6.5, 2.5),
        'ldm': (2, 2), 'cdm': (3.5, 2), 'rdm': (5, 2),
        'lwb': (0.5, 1.5), 'lb': (1.5, 1.5), 'lcb': (2.5, 1.5), 'cb': (3.5, 1.5), 'rcb': (4.5, 1.5), 'rb': (5.5, 1.5), 'rwb': (6.5, 1.5),
        'gk': (3.5, 0.5)
    }

    for pos, (x, y) in position_grid.items():
        rating = position_ratings.get(pos, 0)
        color = 'green' if rating >= 80 else 'blue' if rating >= 70 else 'orange' if rating >= 60 else 'red'
        label = position_labels.get(pos.upper(), pos.upper())
        ax.text(x, y, f"{label}\n{rating}", color='white', ha='center', va='center',
                bbox=dict(facecolor=color, edgecolor='white', boxstyle='round,pad=0.3'))

    plt.title(f"{best_match} (Temp. {selected_version})\n {selected_team}", fontsize=16, color='white', fontweight='bold')
    plt.show()
# Llamar a la funcin con fondo verde csped y sin imagen
plot_player_positions_with_grass_background_only("belotti", "2017", promesas)
```

```python
position_to_column = {
    'lw': 'extr_izq',
    'lf': 'seg_del_izq',
    'cf': 'media_punta',
    'rf': 'seg_del_dch',
    'rw': 'extr_dch',
    'ls': 'del_izq',
    'st': 'del',
    'rs': 'del_dch',
    'lam': 'med_cen_of_izq',
    'cam': 'med_cen_of',
    'ram': 'med_cen_of_dch',
    'lm': 'med_izq',
    'lcm': 'med_cen_izq',
    'cm': 'med_cen',
    'rcm': 'med_cen_dch',
    'rm': 'med_dch',
    'ldm': 'med_def_izq',
    'cdm': 'med_cen_def',
    'rdm': 'med_cen_def_dch',
    'lwb': 'carr_izq',
    'lb': 'lat_izq',
    'lcb': 'def_cen_izq',
    'cb': 'def_cen',
    'rcb': 'def_cen_dch',
    'rb': 'lat_dch',
    'rwb': 'carr_dch',
    'gk': 'portero'
}
position_labels = {
    'GK': 'POR','RWB': 'CAD','RB': 'LAD','CB': 'CEN','RCB': 'CDE','LCB': 'CIZ','LB': 'LIZ','LWB': 'CAI','RM': 'MED',
    'CM': 'MEC','LM': 'MEI','CDM': 'MCD','CAM': 'MCO','RF': 'SDD','CF': 'MPN','LF': 'SDI','RW': 'EXD','ST': 'DC',
    'LW': 'EXI','LS': 'DEI','RS': 'DED','LAM': 'MOI','RAM': 'MOD','LCM': 'MCI','RCM': 'MCD','LDM': 'MDI','RDM': 'MDD'}

def plot_player_positions_with_grass_background_only(player_name, selected_version, promesas):
    # Normalizar los nombres para mejorar la coincidencia
    promesas['nombre_normalizado'] = promesas['nombre_jugador'].str.lower()
    player_name_normalized = player_name.lower()

    # Filtrar jugadores que coincidan con el nombre
    name_parts = player_name_normalized.split()
    player_data = promesas[
        promesas['nombre_normalizado'].apply(lambda x: all(part in x for part in name_parts))
    ]
    
    if player_data.empty:
        print(f"No se encontr ningn jugador con coincidencia para '{player_name}'.")
        return

    # Mostrar las versiones disponibles solo para este jugador
    available_versions = player_data['version_x'].unique()
    print(f"Versiones disponibles para '{player_name}':")
    for version in available_versions:
        team = player_data[player_data['version_x'] == version]['nombre_club_y'].iloc[0]
        print(f"- {version} (Equipo: {team})")
    
    if str(selected_version) not in map(str, available_versions):
        print(f"La versin seleccionada '{selected_version}' no est disponible para '{player_name}'.")
        return

    # Filtrar el jugador por la versin seleccionada
    player_data = player_data[player_data['version_x'].astype(str) == str(selected_version)]
    
    # Usar el primer jugador encontrado
    best_match = player_data.iloc[0]['nombre_jugador']
    selected_team = player_data.iloc[0]['nombre_club_y'].title()
    print(f"Jugador encontrado: {best_match}, Versin: {selected_version}, Equipo: {selected_team}")
    
    # Extraer las posiciones y calificaciones correspondientes
    position_ratings = {}
    for pos, col in position_to_column.items():
        if col in player_data.columns:
            position_ratings[pos] = int(player_data.iloc[0][col]) if not pd.isna(player_data.iloc[0][col]) else 0
        else:
            position_ratings[pos] = 0  # Valor por defecto si la columna no existe

    # Crear el grfico de posiciones
    fig, ax = plt.subplots(figsize=(5, 5))
    fig.patch.set_facecolor('#2E8B57')  # Verde csped
    ax.set_facecolor('#2E8B57')  # Verde csped
    ax.set_xlim(0, 7)
    ax.set_ylim(0, 5)
    ax.axis('off')

    # Dibujar las lneas del campo de ftbol
    field_lines = [
        [(0, 0), (7, 0)],  # Lnea de gol inferior
        [(0, 5), (7, 5)],  # Lnea de gol superior
        [(0, 0), (0, 5)],  # Lnea lateral izquierda
        [(7, 0), (7, 5)],  # Lnea lateral derecha
        [(3.5, 0), (3.5, 5)],  # Lnea central
        [(1.5, 0), (1.5, 1)],  # rea de penal izquierda
        [(5.5, 0), (5.5, 1)],  # rea de penal derecha
        [(1.5, 4), (1.5, 5)],  # rea de penal superior izquierda
        [(5.5, 4), (5.5, 5)],  # rea de penal superior derecha
        [(1.5, 1), (5.5, 1)],  # Lnea de penal inferior
        [(1.5, 4), (5.5, 4)],  # Lnea de penal superior
    ]
    for line in field_lines:
        ax.plot(*zip(*line), color='white', linewidth=1.5)

    # Dibujar el crculo central
    center_circle = plt.Circle((3.5, 2.5), 0.75, color='white', fill=False, linewidth=1.5)
    ax.add_patch(center_circle)

    # Colores basados en las calificaciones
    position_grid = {
        'lw': (0.5, 4), 'lf': (1.5, 4), 'cf': (3.5, 3.5), 'rf': (5.5, 4), 'rw': (6.5, 4),
        'ls': (1, 3.5), 'st': (3.5, 4), 'rs': (6, 3.5),
         'cam': (3.5, 3), 
        'lm': (0.5, 2.5), 'lcm': (2, 2.5), 'cm': (3.5, 2.5), 'rcm': (5, 2.5), 'rm': (6.5, 2.5),
        'ldm': (2, 2), 'cdm': (3.5, 2), 'rdm': (5, 2),
        'lwb': (0.5, 1.5), 'lb': (1.5, 1.5), 'lcb': (2.5, 1.5), 'cb': (3.5, 1.5), 'rcb': (4.5, 1.5), 'rb': (5.5, 1.5), 'rwb': (6.5, 1.5),
        'gk': (3.5, 0.5)
    }

    for pos, (x, y) in position_grid.items():
        rating = position_ratings.get(pos, 0)
        color = 'green' if rating >= 80 else 'blue' if rating >= 70 else 'orange' if rating >= 60 else 'red'
        label = position_labels.get(pos.upper(), pos.upper())
        ax.text(x, y, f"{label}\n{rating}", color='white', ha='center', va='center',
                bbox=dict(facecolor=color, edgecolor='white', boxstyle='round,pad=0.3'))

    plt.title(f"{best_match} (Temp. {selected_version})\n {selected_team}", fontsize=16, color='white', fontweight='bold')
    plt.show()
# Llamar a la funcin con fondo verde csped y sin imagen
plot_player_positions_with_grass_background_only("lautaro martinez", "2019", promesas)
```

# Top Aceleracion

```python
unique_top_acceleration = promesas.loc[
    promesas.groupby('nombre_jugador')['aceleracion'].idxmax(),
    ['version_x', 'nombre_jugador', 'aceleracion']
]

# Sort by acceleration and select the top 10
top_10_acceleration_no_duplicates = unique_top_acceleration.nlargest(10, 'aceleracion')

# Generate color map using the 'viridis' palette
viridis = cm.get_cmap('viridis', len(top_10_acceleration_no_duplicates))
colors = viridis(range(len(top_10_acceleration_no_duplicates)))

# Plot horizontal bar chart
plt.figure(figsize=(10, 5))
plt.barh(
    top_10_acceleration_no_duplicates['nombre_jugador'],
    top_10_acceleration_no_duplicates['aceleracion'],
    color=colors
)
plt.ylabel('Jugadores', fontsize=12)  # Axis label for players
plt.xlabel('Aceleracin', fontsize=12)  # Axis label for acceleration
plt.title('Top - Aceleracin', fontsize=14, fontweight='bold')  # Bold for the title
plt.gca().invert_yaxis()  # Invert y-axis to show the highest values at the top
plt.tight_layout()

# Bold the player names by adjusting the tick labels
plt.yticks(fontweight='bold')

# Show the plot
plt.show()
```

# Top Ritmo

```python
# Extract relevant columns and drop duplicates
promesas_df = promesas[['nombre_jugador', 'version_x', 'ritmo']].drop_duplicates()

# Calculate the average 'ritmo' per player and version
avg_ritmo = promesas_df.groupby(['nombre_jugador', 'version_x'], as_index=False)['ritmo'].mean()

# Get the top 10 players based on average 'ritmo'
top_10_avg_ritmo = avg_ritmo.groupby('nombre_jugador')['ritmo'].mean().nlargest(10).reset_index()

# Extract data for visualization
top_10_players = top_10_avg_ritmo['nombre_jugador']
avg_ritmo_values = top_10_avg_ritmo['ritmo']

# Generate color map using the 'viridis' palette
viridis = cm.get_cmap('viridis', len(top_10_avg_ritmo))
colors = viridis(range(len(top_10_avg_ritmo)))

# Create the horizontal bar chart
plt.figure(figsize=(10, 5))
plt.barh(top_10_players, avg_ritmo_values, color=colors)
plt.xlabel('Ritmo', fontsize=12)
plt.ylabel('Jugador', fontsize=12)
plt.title('Top - Ritmo', fontsize=14, fontweight='bold')
plt.gca().invert_yaxis()  # Invert y-axis to show the highest value at the top

# Customize player names in bold
plt.yticks(fontweight='bold')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

# Top Posicion

```python
# Extraer y limpiar la columna de posiciones
positions_series = promesas['posiciones'].dropna().str.split(',').explode()
cleaned_positions_series = positions_series.str.strip()

# Contar las posiciones combinando duplicados
cleaned_summed_positions = (
    cleaned_positions_series.value_counts()
    .groupby(level=0)
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

# Crear los colores utilizando el esquema viridis
colors = viridis([i / len(cleaned_summed_positions) for i in range(len(cleaned_summed_positions))])

# Crear el grfico de columnas
plt.figure(figsize=(10, 5))
plt.bar(cleaned_summed_positions.index, cleaned_summed_positions.values, color=colors)
plt.title('Top 10 Posiciones', fontsize=16)
plt.xlabel('Posicin', fontsize=14)
plt.ylabel('Cantidad', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Mostrar el grfico
plt.show()
```

# Nacionalidades por Liga

```python
promesas['nombre_liga_y'] = promesas['nombre_liga_y'].str.lower()
ligas_interes = ["premier league", "la liga", "serie a", "bundesliga", "ligue 1"]
data_ligas = promesas[promesas['nombre_liga_y'].isin(ligas_interes)]

# Calcular las 6 nacionalidades ms recurrentes para cada liga
resultados = {}
for liga in ligas_interes:
    data_liga = data_ligas[data_ligas['nombre_liga_y'] == liga]
    top_nacionalidades = data_liga['nacionalidad'].value_counts().head(6)
    resultados[liga.title()] = top_nacionalidades

# Convertir resultados en un DataFrame
ligas_nacionalidades_df = pd.DataFrame(resultados)

# Obtener colores desde la paleta viridis
first_colors = viridis(np.linspace(0, 1, 6))

# Crear grficos
for idx, liga in enumerate(ligas_nacionalidades_df.columns):
    top_nacionalidades = ligas_nacionalidades_df[liga].dropna()
    num_sectors = len(top_nacionalidades)
    colors = first_colors[:num_sectors]
    
    plt.figure(figsize=(5, 5))
    wedges, texts, autotexts = plt.pie(
        top_nacionalidades,
        labels=top_nacionalidades.index,
        autopct='%1.1f%%',
        startangle=90,
        colors=colors,
        textprops={'color': 'black', 'weight': 'bold'}
    )
    
    # Ajustes especficos para "La Liga"
    if "la liga" in liga.lower():
        for text in texts:
            if text.get_text() == "Uruguay":
                x, y = text.get_position()
                text.set_position((x - 0.2, y))  # Mover un poco a la izquierda
    
    # Configurar colores de porcentajes segn el fondo
    for wedge, autotext in zip(wedges, autotexts):
        r, g, b, _ = wedge.get_facecolor()
        if (r > 0.8 and g > 0.8 and b < 0.2) or (g > 0.7 and r < 0.5 and b < 0.5):  # Amarillo o verde
            autotext.set_color('black')
        else:
            autotext.set_color('white')
        autotext.set_weight('bold')
    
    # Ajustar el ttulo
    nacion_liga = promesas[promesas['nombre_liga_y'] == liga.lower()]['nacion_liga'].unique()
    nacion_liga_str = nacion_liga[0].title() if len(nacion_liga) > 0 else "Desconocida"
    if idx == 3:  # Cuarto grfico (ndice 3)
        nacion_liga_str = "Alemania"
    plt.title(f'Nacionalidades - {liga} ({nacion_liga_str})', fontweight='bold')
    plt.show()
```
